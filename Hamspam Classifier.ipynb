```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv("./datasets/hamspam.tsv",sep="\t")
```

```python
dataset.head()
```

```python
dataset.columns=["Labels","Content"]
```

```python
dataset.loc[dataset["Labels"]=="spam","Labels"]=1
```

```python
dataset.loc[dataset["Labels"]=="ham","Labels"]=0
```

```python
import string
```

```python
string.punctuation
```

```python
def messagecleaner(message):
    nonpunc=[]
    for char in message:
        if char not in string.punctuation:
            nonpunc.append(char)
    nonpunc="".join(nonpunc)
    return nonpunc
```

```python
dataset["Content"]=dataset["Content"].apply(messagecleaner)
```
